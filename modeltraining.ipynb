```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Algerian_forest_fires_cleaned_dataset.csv")
```

```python
df.head()
```

```python
# Droping day,month,year
df.drop(['day','month','year'],inplace=True,axis=1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df["Classes"].value_counts()
```

```python
## Independent and dependent features
X=df.drop('FWI',axis=1) # independent feature
y=df['FWI'] #dependent feature
```

```python
X.head()
```

```python
y.head()
```

```python
# Train Test Split
from  sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42 )
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# corerelation between features
# Correlation threshold is a cut-off value used to select features that have strong correlation with the target variable.
```

## Coreralation between featurres

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                col_name=corr_matrix.columns[i]
                col_corr.add(col_name)

    return col_corr

                
```

```python
#threshold--domain expertise
corr_features=correlation(X_train,0.85)
```

```python
## Drop feature when coreraltion is more than 0.85
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
```

```python
X_train
```

## Feature Scaling AND Standarization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

## Box plots after Scaling or Standarization

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("X_train before Scaling")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
linreg=LinearRegression() # our model
linreg.fit (X_train_scaled,y_train)
y_pred=linreg.predict(X_test_scaled)
mae=median_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)

print("MAE:",mae)
print("R2_score:",score)

```

```python
## scatter plot
plt.scatter(y_test,y_pred) 
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso() # our model
lasso.fit (X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=median_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)

print("MAE:",mae)
print("R2_score:",score)


 
```

```python
# r2 score increases that means either the model is overfitting
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge() # our model
ridge.fit (X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mae=median_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)

print("MAE:",mae)
print("R2_score:",score)

```

## Elastic Net Regression

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
elastic=ElasticNet() # our model
elastic.fit (X_train_scaled,y_train)
y_pred=elastic.predict(X_test_scaled)
mae=median_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)

print("MAE:",mae)
print("R2_score:",score)

```
